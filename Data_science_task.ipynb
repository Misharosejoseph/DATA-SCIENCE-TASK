TASK 1: RFM CUSTOMER SEGMENTATION

```python
rfm_base = pd.read_sql("""
SELECT
    c.customer_unique_id,
    MAX(o.order_delivered_customer_date) AS last_order_date,
    COUNT(DISTINCT o.order_id) AS frequency,
    SUM(p.payment_value) AS monetary
FROM olist_orders_dataset o
JOIN olist_customers_dataset c
    ON o.customer_id = c.customer_id
JOIN olist_payments_dataset p
    ON o.order_id = p.order_id
WHERE o.order_status = 'delivered'
GROUP BY c.customer_unique_id
""", conn)

rfm_base.head()
```

CREATE R, F, M SCORES (NTILE)
Logic

Recency  latest date = higher score

Frequency  more orders = higher score

Monetary  more spend = higher score

```python
rfm_scores = pd.read_sql("""
SELECT *,
    NTILE(5) OVER (ORDER BY last_order_date DESC) AS r_score,
    NTILE(5) OVER (ORDER BY frequency) AS f_score,
    NTILE(5) OVER (ORDER BY monetary) AS m_score
FROM (
    SELECT
        c.customer_unique_id,
        MAX(o.order_delivered_customer_date) AS last_order_date,
        COUNT(DISTINCT o.order_id) AS frequency,
        SUM(p.payment_value) AS monetary
    FROM olist_orders_dataset o
    JOIN olist_customers_dataset c
        ON o.customer_id = c.customer_id
    JOIN olist_payments_dataset p
        ON o.order_id = p.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY c.customer_unique_id
) t
""", conn)

rfm_scores.head()

```

CHECK SCORES

```python
rfm_scores[['r_score','f_score','m_score']].describe()
```

CREATE CUSTOMER SEGMENTS

Segmentation Logic
Segment	Condition
Champions	R 4, F 4, M 4
Loyal	R 3, F 3
Potential Loyalists	R 4, F 2
At Risk	R 2, F 3
Lost	Others

```python
rfm_final = pd.read_sql("""
SELECT *,
    CASE
        WHEN r_score >= 4 AND f_score >= 4 AND m_score >= 4 THEN 'Champions'
        WHEN r_score >= 3 AND f_score >= 3 THEN 'Loyal'
        WHEN r_score >= 4 AND f_score <= 2 THEN 'Potential Loyalists'
        WHEN r_score <= 2 AND f_score >= 3 THEN 'At Risk'
        ELSE 'Lost'
    END AS customer_segment
FROM (
    SELECT *,
        NTILE(5) OVER (ORDER BY last_order_date DESC) AS r_score,
        NTILE(5) OVER (ORDER BY frequency) AS f_score,
        NTILE(5) OVER (ORDER BY monetary) AS m_score
    FROM (
        SELECT
            c.customer_unique_id,
            MAX(o.order_delivered_customer_date) AS last_order_date,
            COUNT(DISTINCT o.order_id) AS frequency,
            SUM(p.payment_value) AS monetary
        FROM olist_orders_dataset o
        JOIN olist_customers_dataset c
            ON o.customer_id = c.customer_id
        JOIN olist_payments_dataset p
            ON o.order_id = p.order_id
        WHERE o.order_status = 'delivered'
        GROUP BY c.customer_unique_id
    ) base
) scores
""", conn)

rfm_final.head()
```

TASK 2: REVENUE LEAKAGE DETECTION

Objective

Check if order payment  order item value
Flag orders where difference > 2%

STEP 1: Calculate EXPECTED ORDER VALUE

```python
expected_value = pd.read_sql("""
SELECT
    order_id,
    SUM(price + freight_value) AS expected_value
FROM olist_order_items_dataset
GROUP BY order_id
""", conn)

expected_value.head()
```

STEP 2: Calculate PAID ORDER VALUE

```python
paid_value = pd.read_sql("""
SELECT
    order_id,
    SUM(payment_value) AS paid_value
FROM olist_payments_dataset
GROUP BY order_id
""", conn)

paid_value.head()

```

STEP 3: Compare & FLAG MISMATCHES

Business Rule

Flag if difference > 2%

```python
revenue_leakage = pd.read_sql("""
SELECT
    e.order_id,
    e.expected_value,
    p.paid_value,
    ROUND((p.paid_value - e.expected_value) / e.expected_value, 4) AS diff_pct
FROM (
    SELECT order_id, SUM(price + freight_value) AS expected_value
    FROM olist_order_items_dataset
    GROUP BY order_id
) e
JOIN (
    SELECT order_id, SUM(payment_value) AS paid_value
    FROM olist_payments_dataset
    GROUP BY order_id
) p
ON e.order_id = p.order_id
WHERE ABS((p.paid_value - e.expected_value) / e.expected_value) > 0.02
""", conn)

revenue_leakage.head()
```

TASK 3: FUNNEL CONVERSION ANALYSIS

STEP 1: CREATE FUNNEL COUNTS

```python
funnel_counts = pd.read_sql("""
WITH base AS (
    SELECT
        o.order_id,
        o.order_status,
        r.review_id
    FROM olist_orders_dataset o
    LEFT JOIN olist_order_reviews_dataset r
        ON o.order_id = r.order_id
)

SELECT 'Orders Placed' AS stage, COUNT(DISTINCT order_id) AS count, 1 AS step FROM base
UNION ALL
SELECT 'Orders Paid', COUNT(DISTINCT order_id), 2
FROM base WHERE order_status IN ('paid','shipped','delivered')
UNION ALL
SELECT 'Orders Shipped', COUNT(DISTINCT order_id), 3
FROM base WHERE order_status IN ('shipped','delivered')
UNION ALL
SELECT 'Orders Delivered', COUNT(DISTINCT order_id), 4
FROM base WHERE order_status = 'delivered'
UNION ALL
SELECT 'Orders Reviewed', COUNT(DISTINCT order_id), 5
FROM base WHERE review_id IS NOT NULL
ORDER BY step
""", conn)

funnel_counts
```

STEP 2: ADD CONVERSION & DROP-OFF %

```python
funnel_analysis = pd.read_sql("""
WITH funnel AS (
    SELECT 'Orders Placed' AS stage, COUNT(DISTINCT order_id) AS cnt, 1 AS step
    FROM olist_orders_dataset
    UNION ALL
    SELECT 'Orders Paid', COUNT(DISTINCT order_id), 2
    FROM olist_orders_dataset
    WHERE order_status IN ('paid','shipped','delivered')
    UNION ALL
    SELECT 'Orders Shipped', COUNT(DISTINCT order_id), 3
    FROM olist_orders_dataset
    WHERE order_status IN ('shipped','delivered')
    UNION ALL
    SELECT 'Orders Delivered', COUNT(DISTINCT order_id), 4
    FROM olist_orders_dataset
    WHERE order_status = 'delivered'
    UNION ALL
    SELECT 'Orders Reviewed', COUNT(DISTINCT r.order_id), 5
    FROM olist_order_reviews_dataset r
)

SELECT
    stage,
    cnt AS orders,
    ROUND(100.0 * cnt / LAG(cnt) OVER (ORDER BY step), 2) AS conversion_pct,
    ROUND(100 - (100.0 * cnt / LAG(cnt) OVER (ORDER BY step)), 2) AS dropoff_pct
FROM funnel
ORDER BY step
""", conn)

funnel_analysis
```

TASK 4: COHORT RETENTION ANALYSIS

Objective

Understand how long customers keep coming back after their first purchase.

STEP 1: Find EACH CUSTOMERS FIRST PURCHASE MONTH

Logic

Only delivered orders

First delivered order = cohort start

```python
first_purchase = pd.read_sql("""
SELECT
    c.customer_unique_id,
    MIN(o.order_purchase_timestamp) AS first_purchase_date,
    STRFTIME('%Y-%m', MIN(o.order_purchase_timestamp)) AS cohort_month
FROM olist_orders_dataset o
JOIN olist_customers_dataset c
    ON o.customer_id = c.customer_id
WHERE o.order_status = 'delivered'
GROUP BY c.customer_unique_id
""", conn)

first_purchase.head()
```

STEP 2: Attach ALL ORDERS TO COHORTS

```python
cohort_orders = pd.read_sql("""
SELECT
    c.customer_unique_id,
    STRFTIME('%Y-%m', o.order_purchase_timestamp) AS order_month,
    f.cohort_month,
    (
        (CAST(SUBSTR(STRFTIME('%Y-%m', o.order_purchase_timestamp),1,4) AS INT) -
         CAST(SUBSTR(f.cohort_month,1,4) AS INT)) * 12 +
        (CAST(SUBSTR(STRFTIME('%m', o.order_purchase_timestamp),1,2) AS INT) -
         CAST(SUBSTR(f.cohort_month,6,2) AS INT))
    ) AS month_number
FROM olist_orders_dataset o
JOIN olist_customers_dataset c
    ON o.customer_id = c.customer_id
JOIN (
    SELECT
        c.customer_unique_id,
        STRFTIME('%Y-%m', MIN(o.order_purchase_timestamp)) AS cohort_month
    FROM olist_orders_dataset o
    JOIN olist_customers_dataset c
        ON o.customer_id = c.customer_id
    WHERE o.order_status = 'delivered'
    GROUP BY c.customer_unique_id
) f
ON c.customer_unique_id = f.customer_unique_id
WHERE o.order_status = 'delivered'
""", conn)

cohort_orders.head()
```

STEP 3: BUILD RETENTION TABLE (MONTH 06)

```python
cohort_retention = pd.read_sql("""
SELECT
    cohort_month,
    month_number,
    COUNT(DISTINCT customer_unique_id) AS customers
FROM (
    SELECT
        c.customer_unique_id,
        STRFTIME('%Y-%m', o.order_purchase_timestamp) AS order_month,
        STRFTIME('%Y-%m', MIN(o.order_purchase_timestamp)
            OVER (PARTITION BY c.customer_unique_id)) AS cohort_month,
        (
            (CAST(STRFTIME('%Y', o.order_purchase_timestamp) AS INT) -
             CAST(STRFTIME('%Y', MIN(o.order_purchase_timestamp)
                 OVER (PARTITION BY c.customer_unique_id)) AS INT)) * 12 +
            (CAST(STRFTIME('%m', o.order_purchase_timestamp) AS INT) -
             CAST(STRFTIME('%m', MIN(o.order_purchase_timestamp)
                 OVER (PARTITION BY c.customer_unique_id)) AS INT))
        ) AS month_number
    FROM olist_orders_dataset o
    JOIN olist_customers_dataset c
        ON o.customer_id = c.customer_id
    WHERE o.order_status = 'delivered'
)
WHERE month_number BETWEEN 0 AND 6
GROUP BY cohort_month, month_number
ORDER BY cohort_month, month_number
""", conn)

cohort_retention
```

TASK 5: SELLER PROFITABILITY RANKING

Objective

Identify sellers with high revenue but low profit.

Definitions (VERY IMPORTANT)

Revenue = SUM(price)

Profit = SUM(price  freight_value)

Profit Margin = profit / revenue

STEP 1: CALCULATE SELLER REVENUE & PROFIT

```python
seller_perf = pd.read_sql("""
SELECT
    seller_id,
    SUM(price) AS revenue,
    SUM(price - freight_value) AS profit
FROM olist_order_items_dataset
GROUP BY seller_id
""", conn)

seller_perf.head()
```

```python
seller_perf.shape
```

```python
seller_perf.describe()
```

STEP 3: ADD PROFIT MARGIN & RANKING

```python
seller_ranking = pd.read_sql("""
SELECT
    seller_id,
    revenue,
    profit,
    ROUND(profit * 1.0 / revenue, 4) AS profit_margin,
    RANK() OVER (ORDER BY revenue DESC) AS revenue_rank,
    RANK() OVER (ORDER BY profit * 1.0 / revenue ASC) AS low_margin_rank
FROM (
    SELECT
        seller_id,
        SUM(price) AS revenue,
        SUM(price - freight_value) AS profit
    FROM olist_order_items_dataset
    GROUP BY seller_id
)
""", conn)

seller_ranking.head()
```

STEP 4: FIND HIGH-REVENUE, LOW-PROFIT SELLERS

```python
risk_sellers = seller_ranking[
    (seller_ranking['revenue_rank'] <= 100) &
    (seller_ranking['profit_margin'] < 0.05)
]

risk_sellers.head()
```
